```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
df = pd.read_excel('Final preprocessing dataset.xlsx')
df.head()
```

# EDA

```python
df.columns = df.iloc[0] # set the first row as column name
df = df.iloc[1:] # remove the first row
df.head()
```

```python
print("The dataset has {} rows and {} columns.".format(*df.shape), "\n")
print("It contains {} duplicates.".format(df.duplicated().sum()), "\n")
print("It contains {} NaN.".format(df.isnull().sum().sum()))
```

```python
df.info()
```

## Visualizations

```python
# Annual GDP Prices (Ksh Million)
x = df['Year']
y = df['Real GDP prices (Ksh Million)']
plt.figure(figsize=(10,5))
plt.plot(x,y)
plt.title('Annual GDP Prices (Ksh Million)')
plt.xlabel('Year')
plt.ylabel('Real GDP prices (Ksh Million)')
plt.show();
```

```python
# Real v Nominal GDP
sns.set_style('darkgrid')
plt.figure(figsize=(10,5))
sns.barplot(x='Year', y='Real GDP prices (Ksh Million)', data=df, color='blue', label='Real GDP')
sns.barplot(x='Year', y='Nominal GDP prices (Ksh Million)', data=df, color='red', label='Nominal GDP')
plt.xticks(rotation=90)
plt.xlabel('Year')
plt.ylabel('GDP')
plt.title('Real v Nominal GDP')
plt.legend(['Real GDP', 'Nominal GDP'])
plt.show();
```

```python
# Annual GDP growth rate (%)
plt.figure(figsize=(10,5))
sns.lineplot(x='Year', y='Annual GDP growth (%)', data=df, color='blue')
plt.xticks(rotation=90)
plt.xlabel('Year')
plt.ylabel('GDP growth rate (%)')
plt.title('Annual GDP growth rate (%)')
plt.show();
```

```python
# Unemployment rate v Labour force
plt.figure(figsize=(10,5))
plt.plot(df['Year'], df['Unemployment Rate (%)'], color='blue', label='Unemployment rate (%)')
plt.bar(df['Year'], df['Labour Force'], color='red', label='Labour force')
plt.legend(['Unemployment rate (%)', 'Labour force'])
plt.xticks(rotation=90)
plt.xlabel('Year')
plt.ylabel('Unemployment rate (%)')
plt.title('Unemployment rate v Labour force')
plt.show();
```

```python
# Plot the unemployment rate and labour force on two different y-axes
fig, ax1 = plt.subplots(figsize=(10,5))
ax1.set_xlabel('Year')
ax1.set_ylabel('Labour force', color='red')
ax1.plot(df['Year'], df['Labour Force'], color='red')
ax1.tick_params(axis='y', labelcolor='red')
ax2 = ax1.twinx()
ax2.set_ylabel('Unemployment rate', color='blue')
ax2.plot(df['Year'], df['Unemployment Rate (%)'], color='blue')
ax2.tick_params(axis='y', labelcolor='blue')
plt.title('Unemployment rate v Labour force')
fig.tight_layout()
plt.show();
```

```python
infl_gdp = df[['Year', 'Inflation Rate', 'Annual GDP growth (%)']]

# Plot the inflation rate and GDP growth rate on two different y-axes
fig, ax1 = plt.subplots(figsize=(10,5))
ax1.set_xlabel('Year')
ax1.set_ylabel('Inflation rate', color='red')
ax1.plot(infl_gdp['Year'], infl_gdp['Inflation Rate'], color='red')
ax1.tick_params(axis='y', labelcolor='red')
ax2 = ax1.twinx()
ax2.set_ylabel('GDP growth rate', color='blue')
ax2.plot(infl_gdp['Year'], infl_gdp['Annual GDP growth (%)'], color='blue')
ax2.tick_params(axis='y', labelcolor='blue')
plt.title('Inflation rate v GDP growth rate')
fig.tight_layout()
plt.show();
```

```python
plt.figure(figsize=(10,5))
plt.bar(infl_gdp['Year'], infl_gdp['Inflation Rate'], color='red', label='Inflation rate')
plt.plot(infl_gdp['Year'], infl_gdp['Annual GDP growth (%)'], color='blue', label='GDP growth rate')
plt.xlabel('Year')
plt.ylabel('Inflation rate')
plt.title('Inflation rate v GDP growth rate')
plt.legend()
plt.show();
```
